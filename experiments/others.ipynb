# calculate the SEN, PRE, SPE

```python
import numpy as np

# 实际标签值
actual_labels = np.array([1, 0, 1, 1, 0, 0, 1, 0, 1, 0])

# 预测标签值
predicted_labels = np.array([1, 0, 1, 1, 0, 1, 0, 1, 0, 1])

# 计算真正例数（True Positives）
tp = np.sum((actual_labels == 1) & (predicted_labels == 1))

# 计算真负例数（True Negatives）
tn = np.sum((actual_labels == 0) & (predicted_labels == 0))

# 计算假正例数（False Positives）
fp = np.sum((actual_labels == 0) & (predicted_labels == 1))

# 计算假负例数（False Negatives）
fn = np.sum((actual_labels == 1) & (predicted_labels == 0))

# 计算敏感性（Sensitivity）
sensitivity = tp / (tp + fn)

# 计算特异性（Specificity）
specificity = tn / (tn + fp)

# 计算精确度（Precision）
precision = tp / (tp + fp)

# 打印结果
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
print("Precision:", precision)
```

```python
from sklearn import metrics

# 实际标签值
actual_labels = [1, 0, 1, 1, 0, 0, 1, 0, 1, 0]

# 预测标签值
predicted_labels = [1, 0, 1, 1, 0, 1, 0, 1, 0, 1]

# 计算混淆矩阵
cm = metrics.confusion_matrix(actual_labels, predicted_labels)

# 提取混淆矩阵中的各项值
tn, fp, fn, tp = cm.ravel()

# 计算敏感性（Sensitivity）
sensitivity = tp / (tp + fn)

# 计算特异性（Specificity）
specificity = tn / (tn + fp)

# 计算精确度（Precision）
precision = tp / (tp + fp)

# 打印结果
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
print("Precision:", precision)
```
